### Domestic Consumption

It took some detective work, but the interrelationships of the data contained within the domestic USGS categories were figured out. A full breakdown and its proof is found in the cell below.  The data is very clean, with only a couple of null values for the per capita public supplied and self-supplied categories, both of which are categories that didn't end up being used.

```python
# This cell reveals that TP-TotPop (total population) is a combination of the
# self-supplied population (DD-SSPop) plus the population supplied by public
# water (PS-TOPop), Domestic total use (DO-WDelv) is total domestic 
# self-supplied water withdrawals (DO-WFrTo) plus water withdaws from the 
# public supply (DO-PSDel), publicly supplied per capita use (DO-PSPCp) is
# DO-PSDel times 1000 divided by PS-TOPop, and self-supplied per capita use 
# (DO-SSPCp) is DO-WFrTo times 1000 divided by DO-SSPop.

USGS_exploration = pd.read_csv('./data/usco2015v2.0.csv',  header= 1, \
encoding = 'unicode_escape')
USGS_exploration['analysis1'] = USGS_exploration['DO-SSPop'] +  \
USGS_exploration['PS-TOPop']
USGS_exploration['analysis2'] = USGS_exploration['DO-WFrTo'] +  \
USGS_exploration['DO-PSDel']
USGS_exploration['analysis3'] =  (USGS_exploration['DO-PSDel']*1000) \
/ (USGS_exploration['PS-TOPop'])
USGS_exploration['analysis4'] =  (USGS_exploration['DO-WFrTo']*1000) / \
(USGS_exploration['DO-SSPop'])

USGS_Dom_analysis = USGS_exploration[[ 'DO-SSPop', 'PS-TOPop', 'TP-TotPop',\
'analysis1', 'DO-WFrTo', 'DO-PSDel', 'DO-WDelv ', 'analysis2', 'DO-PSPCp',\
'analysis3', 'DO-SSPCp', 'analysis4']]
USGS_Dom_analysis.head()
```

# Modeling

##  Golf Water Consumption

Water consumption of golf courses was modeled in order to help evaluate their impact on drought-prone regions.  Hence, two subsets were created, one with states from the Southern Plains, Intermountain West, and California-Nevada regions, and one that contained all other states in the lower 48.  In 2015, golf courses in drought-prone states consumed 320 million gallons per day and their average rate of consumption was higher (average mean of 1.22 MGal per day per 1000 acres versus .96 Mgal).  The median and the 25% and 75% quartiles where also significantly higher.  

Linear regression analysis was also attempted.  This was done as a way to evaluate the impact of specific variables such as temperature and acreage on the overall level of water consumption for golf courses. Residuals that deviated from a normal distribution remained a problem throughout and Prob(JB) never went above 0.00 until a log scale model was adopted.  In the log scale model, each increase of 1% in acreage was associated with an increase of about 0.95% in water consumption.   A one degree change in max temperature was associated with a 2.36% increase in the consumptive use of water and a one unit increase in PHDI was associated with 12% decrease in consumptive use. The probability of Jarque Bera was 3.71e-27 which was a significant improvement from where the model had been.  R-squared dropped from 0.82 to .747 but this isn't directly comparable due to log scaling.   The probability of Jarque Bera was 3.71e-27 which was a significant improvement from where the model had been.

## Golf Feature Engineering

One feature was created for golf.  The total amount of water consumed for golf courses was divided by the total amount of acreage of golf courses to create the varaible Golf_con-per_1000_acre. 

Two subsets were also created, namely golf_courses and golf_courses_of_wrath.  Drought.gov was used to determine if a golf course belonged in the wrath category.  If a state was in the [Southern Plains](https://www.drought.gov/dews/southern-plains), [Intermountain West](https://www.drought.gov/dews/intermountain-west), or [California-Nevada](https://www.drought.gov/dews/california-nevada) regions, then it was included as those areas are heavily impacted by low water availability.  This is a conservative definiton of wrath considering other regions such as the [Pacific Northwest](https://www.drought.gov/dews/pacific-northwest) and the [Missouri River basin](https://www.drought.gov/drought-status-updates/drought-status-update-missouri-river-basin-9-1-22) could have been included based off recent weather patterns.

```python
# This creates the feature golf_con-per_1000_acre which is the amount
# of water consumed per 1000 acres of golf course.  The units are Mgal/d
# divided by 1000 acres.
golf['golf_con-per_1000_acre'] =  golf['IG-CUsFr'] / golf["IG-IrTot"]
```

## Golf Descriptive Analysis

Golf courses in the lower 48 states consumed a total of 981 million gallons per day of water in 2015.  320 of those gallons were in states included in the golf courses of wrath category and water consumption was higher for those states than it was for other states.  The mean was 1.22 Mgal/d per 1000 acres whereas it was .96 Mgal/d per 1000 acres for other states.  The median and the 25% and 75% quartiles were also significantly higher.  Both datasets were highly skewed, particularly the other states dataset which had extreme outliers.  Micro-irrigation was associated with the lowest amount of water use. ## Domestic Water Withdraw

There were two primary metrics that could be used to evaluate domestic water withdraw.  The total amount of water withdrawn (i.e. net withdraw) was one.  This metric was highly, highly, correlated with the total amount of population in an area. In some ways this was beneficial as a lot of the variation seen in withdraw can be explained by one variable, namely population.  But it was harmful in other ways as it was difficult to evaluate other cofactors.  For example, the percentage of multi-unit homes in an area was also strongly correlated with population and this could lead to the incorrect conclusion that living in multi-unit homes resulted in higher water consumption when really it was the higher population correlated with multi-unit homes that was the likely cause. (As will be shown, the percentage of multi-unit homes was associated with decreased levels of water consumption once population was accounted for).  This being said, net water withdraw was arguably more important from a regional standpoint as it described the total amount of water being withdrawn for domestic use.

The second metric that could be used was per capita water consumption.  As this metric controls for population by evaluating the amount of water being used *per person* living in an area, other factors that may influence the use of water can be more accurately evaluated.  In other words, it helps to avoid a statistical phenomenon called multicollinearity which is the situation where the predictor variables are themselves highly correlated.  In the multi-unit home example, it avoids the cofounding correlation between multi-unit homes and population so that the variable can be evaluated on a 'per person' basis. There were, however, drawbacks to this approach.  By eliminating the most predictive variable, the overall explanatory effect was significantly reduced.  Secondly, counties with small populations matter as much as counties with large populations in such a model which can result in errors when one is concerned with net water withdraw. This being said, per capita water consumption was arguably the best way to evaluate other factors that have an impact on domestic water withdraw.

Given the strengths and limitations of both approaches, it was decided that two models were necessary, namely one that evaluates net water withdraw and one that evaluates per capita water consumption.  

Both models were highly impacted by residuals that were skewed (i.e. not a normal distribution).  Multiple attempts to correct this by dropping outliers or log scaling were made, but it appears that the phenomenon was a part of the data itself. This notebook will only provide a cursory overview of the model development process.  For more in-depth understanding and analysis, please see [Water Withdraw and Consumption Model Development](./Water_Withdraw_and_Consumption_Model_Development.ipynb)

## Colorado River Basin Case Study 

The picture shown at the start of this notebook is of Lake Mead, first in 2001, then in 2015.  The pictures shown above are the same lake, first in 2000, then in 2022, shown from an aerial perspective.  As can be clearly seen, there is less water in 2022 than there was in the early 2000's. But how much less?  What is the rate of depletion? And, if this trend continues, when will it run out? Having answers to such questions are vitally important for the region's economic and standard of living conditions as such factors are conditional on having access to water.  It is not just Lake Mead that is important.  Lake Powell and the other major reservoirs in the Colorado River Basin also play a role.  In order to help address the aforementioned questions, water storage data for all of the larger reservoirs in the Colorado River basin was compiled and analyzed.  The first model was highly conservative and included all of the data since 1937.  It projected a yearly decline of -152,800 Mgal, but the forecast failed the test portion of the train-test split (r-squared of - 3.734) and the 7,081,000 Mgal level observed on 09, 21, 2022 wasn't supposed to be reached until after 2030. A second model that used all data since 2000 fared better once it was subjected to a test.  It forecasted a yearly decline of -790,832 Mgal, had a .44 r-squared once it was tested, and a mean absolute error of 563,562 Mgal.  It too erred by being too conservative.

The analysis of reservoir data provides some answers regarding the depletion of surface water reserves, but it provides little insight into groundwater.  Experts in the field of hydrology assess groundwater availability through the use of two NASA satellites, namely The Global Land Data Assimilation System (GLDAS) and the Gravity Recovery and Climate Experiment (GRACE). Given the [technical nature of groundwater asessments](https://docserver.gesdisc.eosdis.nasa.gov/public/project/hydrology/GRACEGroundwater.pdf), such analyses were deemed beyond the scope of this project.  Instead, a research paper was consulted.  Published in July 2014 in the journal Geophysical Research Letters, [Groundwater depletion during drought threatens future water security of the Colorado River Basin](https://agupubs.onlinelibrary.wiley.com/doi/full/10.1002/2014GL061055) analyzed 9 years of data (December 2004 to November 2013).  They calculated the loss of groundwater to be −5.6 ± 0.4 km3 yr−1. This converts to 1,479,363 Mgal per year.

Combined, surface and groundwater have a net loss of 2,270,195 Mgal per year. These totals were compared against the total amount of consumption and withdraw. While the removal of water consumption from smaller categories such as golf courses can help, the only categories with enough of a footprint to make a significant dent in the numbers are agriculture and domestic.  Even if one ignores aquifer depletion and only focuses on reservoirs, the 790,832 Mgal deficit cannot be reduced to 0 without significant changes to how the nation consumes water that go well beyond eliminating golf courses in drought prone regions.


## Storage Data From Colorado River Basin Reservoirs

### Data Forecasting

The program Facebook Prophet is an open-source data forecasting tool used to analyze time series data.  It is sensitive to the data given to it so if only data from 2000 onward were given, the forecast would have a sharper downward slope than if all of the data since 1937 were given.  In order to be conservative and avoid any accusations of cherry-picking, one of the projections used all of the available data.  That projection had a yearly decline of -152,800 Mgal.  While the r-squared of the train portion of the forecast was high at .976, it failed when it was tested against all of the values since 2021.  It had forecasted a value of 8,324,972 Mgal by 2030 and the Colorado River Basin was already well below that number. (7,081,000 Mgal as of Sep 21, 2022).  In short, the forecast was too conservative.  The second forecast performed better.  It had a .95 r-squared in the train portion, which dropped to .44 once it was tested with the data from 2021 onwards.  It had a mean absolute error of 291,264 Mgal in the train portion and a 563,561 Mgal error in the test portion.  It too erred by being too conservative in its Sep 21, 2022 forecast. 7,872,734 Mgals had been forecasted.  In both models, the lower bound crossed the zero mark by 2030.  In the second one, the lower bound crossed the zero mark in 2025.

```python
# This imports Facebook Prophet and loads the data for
# all of the years since 1937.

from prophet import Prophet
m = Prophet()
m.fit(Col_storage_Sep_2022);
```

```python
# This sets the end of the year 2030 as the end point of the
# forecast and makes predictions.

future1 = m.make_future_dataframe(periods=3023)
forecast1 = m.predict(future1)
forecast1[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

## Conclusions

In the beginning of this notebook, the stakeholder was loosely defined as everyone as everyone has a stake in the depletion of water reserves.  I will now evaluate what this analysis means for various subsets of this stakeholder grouping, starting first with homeowners.

* Population growth in regions impacted by declining water reserves could be beneficial to homeowners in the short-term by providing housing demand that could drive up prices and giving people who want to exit the market people to sell to, but in the long term it is likely detrimental to home equity.  92% of domestic water withdraw can be explained by population alone (or 94% if one excludes Maricopa County).  Hence, any actions or policies that result in greater levels of population in drought-prone regions are likely to accelerate the depletion of already over-utilized water resources.  The data suggests that this will happen at a rate of about 80 gallons per person per day.  How much of that is consumption and how much of it is just withdraw is unknown at this time.  For long term stakeholders, the lack of water availability should be a primary concern as many household appliances are dependent upon water access.

* The irrigation of lawns may be beneficial in the short-term by providing the illusion that nothing is wrong with water access which could help in the selling of a house(s), but the long-term consequences of such actions are likely to be detrimental to home equity.  While the model for per capita domestic water withdraw is weak considering it only explains 0.258 of the variation seen in the data, the p-values for the weather variables are all statistically significant and are among the most important, second only to non-domestic irrigation with fresh surface-water.  The most likely explanation for these correlations is watering lawns or gardens as other domestic uses of water are less likely to be greatly impacted by weather or have an association with other forms of irrigation.  If dry states could reduce their withdraw levels to that of other states, the potential water savings would be considerable.  (Rough estimate for the other states category - 1399 Mgal per day) 

* Long-term, it is best for owners and investors in golf courses to have smaller acreage in regions where water availability is an issue as acreage is the most important variable when it comes to predicting water consumption.   Alternatively, one can use surfaces that don't require watering. Some innovative business leaders have already taken these steps (see above).  Maximum temperature was the next biggest factor so increases in temperature will probably result in more water being used for golf courses.  A one degree change in maximum temperature was associated with a 0.0236 MGal increase in daily irrigation. There was not much data to go on as almost all golf courses use sprinklers, but what data exists suggests micro-irrigation should be used in drought-prone regions.  It is debatable whether such changes are feasible in the short term given that it could be expensive, the changes may run counter to the expectations of one's customer base, and the impact of a single golf course will only delay the end result if other golf courses and other industries/consumers are not also applying water conservation strategies.  In short, golf courses may encounter tragedy of the commons problems.  This will be discussed in a bit.

* Residents and people who purchase products made from raw materials produced in drought-prone regions should be cognizant of reservoir forecasts. A conservative forecast that included all of the reservoir storage data since 1937 for the Colorado River Basin had water declining at yearly rate of -152,800 Mgal and the lower bound crossing the zero mark by 2030.  A more pessimistic (and accurate) forecast that included all of the reservoir storage data since 2000 had water declining at yearly rate of -790.83 Mgal and the lower bound crossing the zero mark in 2025.  In short, it is entirely possible for all of the reservoirs in the Colorado River Basin to be out of water in less than ten years.  While it is also possible for water reserves to increase in volume, this would run counter to the trendline seen since 2000.  For people who work and live in the region, it is advisable to have a plan in place about what to do should the trendline continue to persist in its downward trajectory.  A lack of water availability is likely to be highly disruptive to the region's economy and living conditions. For those people who buy products made from raw materials produced in the region such as food and clothes, it may be advisable to buy some products in advance as scarcity brought about by a lack of water is likely to create shortages and drive-up prices.

It was left out of this analysis because it was a very complex topic given the various crops and regions, but it is the biggest consumer of water in areas where water availability is a concern. ## Discussion Concerning the Tragedy of the Commons

I used to work as an outdoor educator and one of the classes I taught was called "Tod the Cod".  During the class, one would fill a large bowl with Goldfish crackers, group students (usually 5th and 6th graders) into groups of two or three and hand them their fishing boats (small plastic lunch bags).  Then you would tell the students something like "You are all fishermen and your job is to maximize your profit over ten seasons.  If you don't catch at least 5 fish in a season, you will go bankrupt and the bank will reclaim your boat.  Now go fish."  The students would then proceed to grab as many fish as they could in the first round, often leaving the bowl without any fish left.  Sometimes leaving just a few behind.  Then the next round most of the classes that left fish behind would also have an empty fish bowl as the remaining fish didn't reproduce enough to replenish the stock.  Once the bowl was empty, one would have the students dump the goldfish back into the bowl and have them try again.  Of the various times I taught the class, there was only one group of students that didn't have an empty fishbowl by round three during their first attempt.  I was shocked.

The idea behind the class was to teach students about a game theory problem first mentioned by the economist William Forster Lloyd in [Two Lectures on the Checks to Population](https://en.wikisource.org/wiki/Two_Lectures_on_the_Checks_to_Population) and popularized by Garrett Hardin in 1968 in his article "The Tragedy of the Commons".   In the case of Llyod and Hardin, the tragedy was the overgrazing of common fields by cows owned by ranchers.  In the case of Tod the Cod, it was overfishing Goldfish crackers by student fishermen.  In the case of reservoirs and aquifers, it is the depletion of water resources by domestic and commercial interests.  Despite its various manifestations, the cause is the same. As Hardin states in his article, "herein is the tragedy. Each man is locked into a system that compels him to increase his herd without limit – in a world that is limited. Ruin is the destination toward which all men rush, each pursuing his own best interest in a society that believes in the freedom of the commons."   In the case of my students, they couldn't come to a sustained system of profit until they came to an agreement amongst each other because rational self-interest without such an agreement compelled them to destroy the common.  Something similar needs to happen in regards to water resources because, unlike with my students, there isn't an easy way refill the fishbowl once we mess up.  We have only one chance to get it right. 

Solutions to tragedy of the commons problems doesn't necessarily mean turning to the government for guidance.  [Government policies, are, after all, one of the principal causes of the water depletion](https://www.taxpayer.net/energy-natural-resources/federal-subsidies-corn-ethanol-corn-based-biofuels/), and my students never needed me telling them how to solve the problem.  But it does require a sense of collective purpose, fairness, and a belief that those who abuse the common will be held accountable.  In the case of the golf courses mentioned earlier, it may make little sense to pay the expense of converting to micro-irrigation if one knows that nobody else is doing it and therefore the water will run out anyway.  If, on the other hand, most people are using micro-irrigation and all of those who refuse to are subjected to heavy fines, then it becomes sensible to pay for the micro-irrigation, particularly if the costs to convert is less than the fines.  

It is not my place to say precisely how we should collectively solve this problem, but I do feel compelled to share what I am seeing in the data and jumpstart the conversation.  Maybe I am wrong, but the way I see it is that the last twenty years of data indicates that the reservoirs will be close to zero (or possibly below zero) in the next ten years.  How much remains in aquifers is unknown.  That is not a lot of time to make corrective actions and it is better to act sooner than later.
