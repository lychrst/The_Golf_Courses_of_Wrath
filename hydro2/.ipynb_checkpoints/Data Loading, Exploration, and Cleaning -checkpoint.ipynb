##  Exploring USGS 

The U.S. Geological Survey's (USGS's) National Water Use Science Project is a difficult csv to comprehend.  For one, the categories used in the [project are evolving](https://www.usgs.gov/mission-areas/water-resources/science/changes-water-use-categories?qt-science_center_objects=0#qt-science_center_objects) so data from 1995 might not match with data taken from 2015.  Secondly, the descriptions of the definitions used, while useful, are often difficult to comprehend at a deeper level.  A link to 2015 definitions used by the USGS can be found [here](https://www.sciencebase.gov/catalog/file/get/5af3311be4b0da30c1b245d8?f=__disk__7c%2Fd0%2Fda%2F7cd0da1f354d12c5870f0dee7446cb41ad5a013f&transform=1&allowOpen=true#Metadata%20Reference%20Information).  Three principal sources were used to help interpret the USGS data.  The first source is [Estimated Use of Water in the United States Data Dictionary for 1995 Data Files](https://water.usgs.gov/watuse/data/1995/datadict.html).  Unline the 2015 data dictionary, the 1995 version clearly explains the relationships between the data columns similar to what one might see on a IRS 1044 tax form.  For example, the column "Per-capita use, public-supplied, in gallons per day" is column 52 times 1000 divided by column 5.  Some of these same relationshps were maintained in the 2015 analysis.  The second source was a USGS [Water-Use Terminology](https://www.usgs.gov/mission-areas/water-resources/science/water-use-terminology) webpage and the thrid source was the USGS's [Estimated Use of Water in the United States in 2015](https://pubs.usgs.gov/circ/1441/circ1441.pdf) report itself.  Combined, the three sources provide a fairly comprehensive overview of the csv.  

In addition to analyzing the aforementioned literature, exporatory analyses were conducted on the data itself in order to better comprehend the interrelationships.  The following are those analyses, broken down by the categories analyzed -  (get).  Analysis on (get) were not attempted.  Descriptive analyses were also not attempted as that information was already available through the USGS's 2015 report. 

The USGS dataset is quite large, with 3223 rows of county-level data and 141 columns of values.  Almost all of the values are numeric, the exceptions being columns like state and county.

### Domestic Consumption


```python
# This cell reveals that TP-TotPop (total population) is a combination of the self-supplied 
# population (DD-SSPop) plus the population supplied by public warter (PS-TOPop), Domestic
# total use (DO-WDelv) is total domestic self-supplied water withdrawals (DO-WFrTo) plus water 
# withdaws from the public supply (DO-PSDel), publicly supplied per capita use (DO-PSPCp) is
# DO-PSDel times 1000 divided by PS-TOPop, and self-supplied per capita use (DO-SSPCp) is 
# DO-WFrTo times 1000 divided by DO-SSPop.

USGS_exploration = hydro2
USGS_exploration['analysis1'] = USGS_exploration['DO-SSPop'] +  USGS_exploration['PS-TOPop']
USGS_exploration['analysis2'] = USGS_exploration['DO-WFrTo'] +  USGS_exploration['DO-PSDel']
USGS_exploration['analysis3'] =  (USGS_exploration['DO-PSDel']*1000) / (USGS_exploration['PS-TOPop'])
USGS_exploration['analysis4'] =  (USGS_exploration['DO-WFrTo']*1000) / (USGS_exploration['DO-SSPop'])

USGS_Dom_analysis = USGS_exploration[[ 'DO-SSPop', 'PS-TOPop', 'TP-TotPop', 'analysis1', \
'DO-WFrTo', 'DO-PSDel', 'DO-WDelv ', 'analysis2', 'DO-PSPCp', 'analysis3', 'DO-SSPCp', 'analysis4']]
USGS_Dom_analysis.head()
```

```python
# DO-SSPop, PS-TOPop, TP-TotPop, DO-WFrTo, DO-PSDel and DO-WDelv have 0 null values. DO-PSPCp
# has 4 null values and DO-SSPCp has 189 null values.  Everything is stored as a float.
USGS_Dom_analysis.info()
```

## Exploring NOAA

The [NOAA website](https://www.ncei.noaa.gov/access/monitoring/climate-at-a-glance/statewide/mapping/110/phdi/201508/1/value) enables one to download specific csvs of interest, but it returns some interesting results that need to be carefully dealt with.  The value of anything requested such as 2015 precipitation data will return "Value" as the column name.  Thus, if one is doing multiple requests for things like precipitation and maximum temperature, one will need to rename "value" to the variable it represents.  Otherwise the combined dataset would have mulitple columns named value and it would not be immediately apparent what those various columns represent.   Another factor was that I was unable to download data for things like PHDI beyond a month time scale which meant that multiple downloads and merges were necessary.  These factors resulted in a fairly lengthy data cleaning process.

NOAA only contains values for the continguous United States which means Alaska, Hawaii, and Puerto Rico are not included in the dataset.  All of its values are numerical, although the Palmer Hydorological indexes and the heating and cooling days are similar to categorical values as only state-level information was available.

## Cleaning NOAA
ss

```python
# This renames the value columns for the variables I have yearly
# data for so that a merged dataset can be easily interpreted.

precip.rename(columns = {"Value":"Precip"}, inplace =True)
Min_temp.rename(columns = {"Value":"Min_temp"}, inplace =True)
Max_temp.rename(columns = {"Value":"Max_temp"}, inplace =True)
Avg_temp.rename(columns = {"Value":"Avg_temp"}, inplace =True)
cooling_day.rename(columns = {"Value":"Cool_day"}, inplace =True)
heating_day.rename(columns = {"Value": "Heat_day"}, inplace =True)
```

```python
# This uses converts the location ID into a value I can use for a 
# merge by seperating the states abbreviation from the numbers that follow it.

precip['State'] = precip["Location ID"].astype(str).str[:2]
Min_temp['State'] = Min_temp["Location ID"].astype(str).str[:2]
Max_temp['State'] = Max_temp["Location ID"].astype(str).str[:2]
Avg_temp['State'] = Avg_temp["Location ID"].astype(str).str[:2]
```

```python
# This adds the various monthly values into a yearly PHDI value that 
# can be used in subsequent analyses.
Jan2015_PHDI["PHDI"] = Jan2015_PHDI["Value"] + Feb2015_PHDI["Value"] + Mar2015_PHDI["Value"] + Apr2015_PHDI["Value"] + \
May2015_PHDI["Value"] + Jun2015_PHDI["Value"] + Jul2015_PHDI["Value"] + Aug2015_PHDI["Value"] + \
Sep2015_PHDI["Value"] + Oct2015_PHDI["Value"] + Nov2015_PHDI["Value"] + Dec2015_PHDI["Value"]
```

```python
# Same as above, only for PDSI.
Jan2015_PDSI["PDSI"] = Jan2015_PDSI["Value"] + Feb2015_PDSI["Value"] + Mar2015_PDSI["Value"] + \
Apr2015_PDSI["Value"] + May2015_PDSI["Value"] + Jun2015_PDSI["Value"] + Jul2015_PDSI["Value"] + \
Aug2015_PDSI["Value"] + Sep2015_PDSI["Value"] + Oct2015_PDSI["Value"] + Nov2015_PDSI["Value"] + Dec2015_PDSI["Value"]
```

```python
# Same as above, only for PMDI.
Jan2015_PMDI["PMDI"] = Jan2015_PMDI["Value"] + Feb2015_PMDI["Value"] + Mar2015_PMDI["Value"] + \
Apr2015_PMDI["Value"] + May2015_PMDI["Value"] + Jun2015_PMDI["Value"] + Jul2015_PMDI["Value"] + \
Aug2015_PMDI["Value"] + Sep2015_PMDI["Value"] + Oct2015_PMDI["Value"] + Nov2015_PMDI["Value"] + Dec2015_PMDI["Value"]
```

```python
# This drops the unnessary columns.
precip = precip.drop(['Location ID', 'Rank', 'Anomaly (1901-2000 base period)', '1901-2000 Mean'], axis = 1)
Min_temp = Min_temp.drop(['Location ID', 'Rank', 'Anomaly (1901-2000 base period)', '1901-2000 Mean'], axis = 1)
Max_temp = Max_temp.drop(['Location ID', 'Rank', 'Anomaly (1901-2000 base period)', '1901-2000 Mean'], axis = 1)
Avg_temp = Avg_temp.drop(['Location ID', 'Rank', 'Anomaly (1901-2000 base period)', '1901-2000 Mean'], axis = 1)
cooling_day = cooling_day.drop(['Location ID','Rank', 'Anomaly (1901-2000 base period)', '1901-2000 Mean'], axis = 1)
heating_day = heating_day.drop(['Location ID', 'Rank', 'Anomaly (1901-2000 base period)', '1901-2000 Mean'], axis = 1)
PHDI = Jan2015_PHDI.drop(['Location ID', 'Rank', 'Anomaly (1901-2000 base period)', '1901-2000 Mean', "Value"], axis = 1)
PDSI = Jan2015_PDSI.drop(['Location ID', 'Rank', 'Anomaly (1901-2000 base period)', '1901-2000 Mean', "Value"], axis = 1)
PMDI = Jan2015_PMDI.drop(['Location ID', 'Rank', 'Anomaly (1901-2000 base period)', '1901-2000 Mean', "Value" ], axis = 1);
```

```python
# This combines the variables I only have state values for into a single dataframe. 
NOAA1 = pd.merge(PHDI, PDSI, on = ["Location"])
NOAA2 = pd.merge(NOAA1, PMDI, on = ["Location"])
NOAA3 = pd.merge(NOAA2, cooling_day, on = ["Location"])
NOAA4 = pd.merge(NOAA3, heating_day, on = ["Location"])
```

```python
# The handy-dandy state_names dataframe comes into play again.  This time it is 
# being merged with the state values so that that they have abbreviations.
NOAAabr = pd.merge(NOAA4, state_names, left_on=['Location'], right_on = ['state']);
```

```python
# This shows were we are in the data cleaning process.
NOAAabr.head()
```

```python
# This merges the variables I have county values for into a single dataframe.
NOAA5 = pd.merge(precip, Min_temp, on = ['State', "Location"])
NOAA6 = pd.merge(NOAA5, Max_temp, on = ['State', "Location"])
NOAA7 = pd.merge(NOAA6, Avg_temp, on = ['State', "Location"])
```

```python
# This merges everything for NOAA into a single dataframe.
NOAA = pd.merge(NOAAabr, NOAA7, left_on = ["abr"], right_on = ['State'])
```

```python
# Yeah.  It is clean... sort of.
NOAA
```

```python
# This is a simple way to change the column name to county.
NOAA['County'] = NOAA['Location_y']
```

```python
# This drops unneccessay columns.
cleaned_NOAA = NOAA.drop(['Location_x','abr', 'Location_y', 'state'], axis = 1)      
```

```python
# Finally, 'La Salle Parish'is changed to 'LaSalle Parish' so its values aren't lost
# once NOAA is merged with USGS.
cleaned_NOAA['County'] = cleaned_NOAA['County'].str.replace('La Salle Parish','LaSalle Parish')
```
